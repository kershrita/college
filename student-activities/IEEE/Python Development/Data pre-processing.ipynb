```python
import pandas as pd
```

```python
dataset = pd.read_csv("housing.csv")
dataset
```

```python
dataset.dtypes
```

```python
dataset.fillna(dataset.mean(), inplace=True)
```

```python
dataset.describe()
```

```python
label = ["total_rooms", "total_bedrooms", "population", "households", "median_house_value"]
```

```python
Q1 = dataset[label].quantile(0.25)
Q3 = dataset[label].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_limitation = Q1 - 1.5 * IQR
upper_limitation = Q3 + 1.5 * IQR
print(lower_limitation)
print('-'*40)
print(upper_limitation)
```

```python
lower_outlires = (dataset[label] < lower_limitation)
upper_outlires = (dataset[label] > upper_limitation)
```

```python
dataset
```

```python
for i in label:
    dataset = dataset[~(lower_outlires[i] | upper_outlires[i])]
dataset
```
